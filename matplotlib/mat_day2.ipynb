```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(0,10,100)
y1= np.sin(x)
y2= np.cos(x)
y3= np.tan(x)

plt.plot(x,y1,label="sin(x)",color='blue')
plt.plot(x,y2,label="cos(x)",color='red')
plt.plot(x,y3,label="tan(x)",color="green")

plt.xlabel('x')
plt.ylabel('y')
plt.title("Multiple lines in a plot")
plt.legend()
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(0,10,100)
y1= np.sin(x)
y2= np.cos(x)
plt.plot(x, y1, 'b-', x , y2, 'r--')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Multiple lines in single plot call")
plt.legend()
plt.grid()
plt.show()
plt.pause(2)
plt.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(0,10,100)
functions=[(np.sin(x),'sin(x)','blue'),(np.cos(x),'cos(x)','green'),(x**2,'x^2','brown')]
for func,label,color in functions:
    plt.plot(x,func,label=label,color=color)
plt.xlabel('x')
plt.ylabel('y')
plt.title("multiple lines using a loop")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4]
y1= [10,20,30,40]
y2= [30,20,40,10]
y3=[25,35,15,20]
y4=[23,45,23,56]
plt.subplot(2,2,1)  # plt.subplot(3,1,1)
plt.plot(x,y1,color='blue')
plt.title("first plot")
plt.subplot(2,2,2)
plt.plot(x,y2,color='red')
plt.title("second plot")
plt.subplot(2,2,3)
plt.plot(x,y3,color='green')
plt.title("thrid plot")
plt.subplot(2,2,4)
plt.plot(x,y4,color='yellow')
plt.title("fourth plot")

plt.tight_layout()         #'''to over come the overlap situation we use tight_plot  '''
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(10,6))
ax = fig.add_subplot(111,projection='3d')
x=np.array([10,20,30])
y=np.array([10,20,30])
z= np.zeros(3)
dx= dy = np.ones(3)*0.4
dz= np.array([[5,10,15],[10,5,15],[15,20,5]])

for i in range(len(x)):
    for j in range(len(y)):
        ax.bar3d(x[i],y[j],z[i],dx[i],dy[j],dz[i,j],color='red',edgecolor='black')

ax.set_xlabel('X-Axis')
ax.set_ylabel('Y-Axis')
ax.set_zlabel('Z-Axis')
ax.set_title('3D-Plot')
ax.view_init(elev=15,azim=60)
plt.tight_layout()
plt.show()
```
