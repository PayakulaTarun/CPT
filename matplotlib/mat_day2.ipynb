```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(0,10,100)
y1= np.sin(x)
y2= np.cos(x)
y3= np.tan(x)

plt.plot(x,y1,label="sin(x)",color='blue')
plt.plot(x,y2,label="cos(x)",color='red')
plt.plot(x,y3,label="tan(x)",color="green")

plt.xlabel('x')
plt.ylabel('y')
plt.title("Multiple lines in a plot")
plt.legend()
plt.grid()

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x= np.linspace(0,10,100)
y1= np.sin(x)
y2= np.cos(x)
plt.plot(x, y1, 'b-', x , y2, 'r--')
plt.xlabel('x')
plt.ylabel('y')
plt.title("Multiple lines in single plot call")
plt.legend()
plt.grid()
plt.show()
plt.pause(2)
plt.close()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x= np.linspace(0,10,100)
functions=[(np.sin(x),'sin(x)','blue'),(np.cos(x),'cos(x)','green'),(x**2,'x^2','brown')]
for func,label,color in functions:
    plt.plot(x,func,label=label,color=color)
plt.xlabel('x')
plt.ylabel('y')
plt.title("multiple lines using a loop")
plt.legend()
plt.grid()
plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4]
y1= [10,20,30,40]
y2= [30,20,40,10]
y3=[25,35,15,20]
plt.subplot(1,3,1)  # plt.subplot(3,1,1)
plt.plot(x,y1)
plt.title("first plot")
plt.subplot(1,3,2)
plt.plot(x,y2)
plt.title("second plot")
plt.subplot(1,3,3)
plt.plot(x,y3)
plt.title("thrid plot")

plt.show()
```

```python
import matplotlib.pyplot as plt
x= [1,2,3,4]
y1= [10,20,30,40]
y2= [30,20,40,10]
y3=[25,35,15,20]
y4=[23,45,23,56]
plt.subplot(2,2,1)  # plt.subplot(3,1,1)
plt.plot(x,y1,color='blue')
plt.title("first plot")
plt.subplot(2,2,2)
plt.plot(x,y2,color='red')
plt.title("second plot")
plt.subplot(2,2,3)
plt.plot(x,y3,color='green')
plt.title("thrid plot")
plt.subplot(2,2,4)
plt.plot(x,y4,color='yellow')
plt.title("fourth plot")

plt.tight_layout()         #'''to over come the overlap situation we use tight_plot  '''
plt.show()
```
